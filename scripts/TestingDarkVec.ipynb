```python
from pathlib import Path
DATA_FOLDER = Path("/content/drive/MyDrive/DSCI599/data")

assert DATA_FOLDER.exists()
```

```python
import pandas as pd

df_path = DATA_FOLDER / "1464217200.csv"
df = pd.read_csv(df_path)

df.head()
```

```python
import pandas as pd
GT = DATA_FOLDER / "big_gt.csv"

gt = pd.read_csv(GT, header=0)
gt.head()
```

```python
import ipaddress

def conv_ip_intip(ip: str):
  assert type(ip) == str
  return int(ipaddress.ip_address(ip.strip()))

gt['int_ip'] = gt['ip'].apply(conv_ip_intip)

gt.head()
```

```python
left = df.rename(columns={'int_source_ip': 'ip'})
left['ip'] = left['ip'].astype('int')

right = gt[['int_ip', 'class']].rename(columns={'int_ip': 'ip'})

join = left.merge(right,
        on='ip',
        how='inner')
join.loc[~join['class'].isna(), :]
```

```python
join['class'] = join['class'].fillna("Unknown")
join['class'].value_counts()
```
