# Load data as a Heterogenous graph

Since lots of the GCN are built for undirected graphs, let's load the data as a heterogenous graph

Plan:
- Load each source ip -> destination port IF they actually have connection in dataset
- Set everything as source ip feature except int_source_ip and destination
- OHE the destination port to set it as categorical
- make one edge per src_ip and dest_port
- edge feature is protocol

make one edge per tcp(6), udp(17), icmp(1)

For each int_source_ip -> (protocol) -> destination_port, get the mean of everything

This to ensure only one edge per source and destination to avoid complications

```python
grouped_df = proc_df.groupby(['int_source_ip', 'destination_port', 'protocol'])
grouped_df = grouped_df.mean().reset_index(drop=False)
grouped_df.head()
```

## Shallow: MetaPath2Vec

> https://pytorch-geometric.readthedocs.io/en/stable/tutorial/shallow_node_embeddings.html

> https://pytorch-geometric.readthedocs.io/en/stable/generated/torch_geometric.nn.models.MetaPath2Vec.html#torch_geometric.nn.models.MetaPath2Vec

> https://github.com/pyg-team/pytorch_geometric/blob/master/examples/hetero/metapath2vec.py

```python
import torch
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

```python
from torch_geometric.nn import MetaPath2Vec

# similar to
# https://github.com/pyg-team/pytorch_geometric/blob/master/examples/hetero/metapath2vec.py
embedding_dim = 64
walk_length = 10  # lower because it's a bipartite graph
context_size = 3
walks_per_node = 10
num_negative_samples = 5

metapath = data.metadata()[1]
num_nodes_dict = {
    'source_ip': data['source_ip'].num_nodes,
    'dest_ports': data['dest_ports'].num_nodes
}

model = MetaPath2Vec(data.edge_index_dict,
                     embedding_dim,
                     metapath,
                     walk_length,
                     context_size,
                     walks_per_node=walks_per_node,
                     num_negative_samples=num_negative_samples,
                     num_nodes_dict=num_nodes_dict,
                     sparse=False).to(device)
```

Train

```python
from tqdm import tqdm
loader = model.loader(batch_size=256, shuffle=True, num_workers=0)
optimizer = torch.optim.Adam(list(model.parameters()), lr=0.01)

# from
# https://github.com/pyg-team/pytorch_geometric/blob/master/examples/hetero/metapath2vec.py
def train(epoch, log_steps=100, eval_steps=2000):
    model.train()

    total_loss = 0
    for i, (pos_rw, neg_rw) in enumerate(loader):
        optimizer.zero_grad()
        loss = model.loss(pos_rw.to(device), neg_rw.to(device))
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        if (i + 1) % log_steps == 0:
            print(f'Epoch: {epoch}, Step: {i + 1:05d}/{len(loader)}, '
                  f'Loss: {total_loss / log_steps:.4f}')
            total_loss = 0
```

```python
EPOCHS = 7
for epoch in range(0, EPOCHS):
    train(epoch)
```

## INCOMPLETE: GraphSAGE

```python
from torch_geometric.nn import SAGEConv, to_hetero
import torch.nn.functional as F

# from link above
class GNN(torch.nn.Module):
    def __init__(self, hidden_channels, out_channels):
        super().__init__()
        self.conv1 = SAGEConv((-1, -1), hidden_channels)
        self.conv2 = SAGEConv((-1, -1), out_channels)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index).relu()
        x = self.conv2(x, edge_index)
        return x
```

```python
num_classes = data['source_ip'].y.max().item() + 1
num_classes
```
